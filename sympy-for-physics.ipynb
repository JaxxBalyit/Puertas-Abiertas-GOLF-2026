**SymPy** (*symbolic python*) es una librería de Python para matemática simbólica, es decir, que trabaja con expresiones algebraicas en lugar de números concretos. Nos permite:
- Resolver ecuaciones algebraicas y diferenciales
- Calcular límites, derivadas e integrales
- Manipular expresiones matemáticas simbólicas
- Resolver problemas de física analíticamente

Comencemos con los fundamentos básicos de SymPy:

````python
# Importación de librerías
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
````

Sabemos que en [*LáTeX*](https://www.latex-project.org/) se usan comandos para escribir fórmulas matemáticas; de manera similar, se realiza en SymPy.
Para declarar símbolos en LáTeX, utilizamos distintas simbologías, tales como `$...$`, `\[...\]`, entre otros. Ejemplo:
```tex
$\alpha$
\[\beta\]
```
La redenrización del código de arriba sería:
$$
\alpha\\
\beta
$$

## Resolución de Ecuaciones con Sympy
Resolveremos la siguiente ecuación cuadrática:
$$
x^2 - 5x + 6 = 0
$$

````python
# Declaración de la función
y = x**2 - 5*x + 6
y
````

````python
# Llamada al método solve(f, x) [función, variable]
sol = smp.solve(y, x)
sol
````

También podemos resolver sistemas de ecuaciones:
$$
\begin{cases}
a + b = 5 \\
2a - b = 1
\end{cases}
$$

````python
a,b = smp.symbols('a b')
eqn1 = a + b - 5
eqn2 = 2*a - b - 1
smp.solve([eqn1, eqn2], [a, b])
````

## Solución a problemas físicos
Supongamos que tenemos un objeto que se encuentra en caída libre, así como una plataforma que va en ascenso.
$$
h_o(t) = h_0-v_ot-\frac{1}{2}gt^2\\
h_p(t) = v_pt+\frac{1}{2}qt^2
$$
El problema nos pide encontrar la velocidad inicial del objeto para que choque con la plataforma, y que cuando colisionen, los dos objetos tengan la misma velocidad. Se ven involucradas 2 ecuaciones:
$$
h_0(t) = h_p(t)\\
\frac{dh_0}{dt}(t) = -\frac{dh_p}{dt}(t)
$$

````python
t, h0, v0, g, vp, q = smp.symbols('t h_0 v_0 g v_p q', real=True, positive=True)
````

````python
# Definimos las ecuaciones de movimiento y sus derivadas
h0t = h0 - v0*t - smp.Rational(1, 2)*g*t**2
dh0dt = g*t - v0

hpt = vp*t + smp.Rational(1, 2)*q*t**2
dhpdt = vp + q*t

# Definimos las ecuaciones involucradas
eq1 = h0t - hpt
eq2 = dh0dt + dhpdt
````

````python
# Procedemos a resolver las ecuaciones
t_col, v_ini = smp.solve([eq1, eq2], [t, v0])[0]
````

````python
t_col, v_ini
````

**¿Cuál es la velocidad a la que colisionan?**

````python
dh0dt.subs([(t, t_col), (v0, v_ini)]).simplify()
````

````python
dhpdt.subs([(t, t_col), (v0, v_ini)]).simplify()
````

## Cálculo Diferencial e Integral
### Limites
$$
\lim_{x \to 0} \frac{\sin(x)}{x}
$$

Puesto en código:

````python
smp.limit(smp.sin(x)/x, x, 0)
````

### Derivadas
$$
\frac{d}{dx}\left(\frac{1+\sin(x)}{1-\cos(x)}\right)^2
$$

````python
smp.diff(((1+smp.sin(x)) / (1-smp.cos(x)))**2,x)
````

### Antiderivadas
$$
\int \csc(x)\cot(x)dx
$$

````python
smp.integrate(smp.csc(x)*smp.cot(x), x)
````

### Integrales Definidas
$$
\int_0^{\ln(4)} \frac{e^x}{\sqrt{e^{2x}+9}} dx
$$

````python
smp.integrate(smp.exp(x) / smp.sqrt(smp.exp(2*x)+9), (x, 0, smp.log(4)))
````

$$
\int_0^\infty e^{-x^2} dx
$$

````python
smp.integrate(smp.exp(-x**2), (x, 0, smp.oo))
````

## Problemas de valores iniciales
Suponga que tenemos lo siguiente:
$$
\frac{dy}{dt} = 8x+\csc^2(x) \text{ con las condiciones iniciales } y(\pi/2)=-7, \text{Resuleva para }y(x)
$$

````python
integral = smp.integrate(8*x+smp.csc(x)**2, x)
````

````python
integral.subs(x, smp.pi/2)
````

````python
C = - integral.subs(x, smp.pi/2) - 7

# Aplicando lógica
y = integral + C
y
````

````python
y.subs(x, smp.pi/2)
````
